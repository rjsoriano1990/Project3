```python
import os
import pandas as pd
from sqlalchemy import create_engine
from config import post_pass
```

```python
csv_file = os.path.join("Resources/Gas_Prices_Per_Gallon_11-21.csv")
gas_per_gallon_df = pd.read_csv(csv_file)
gas_per_gallon_df.head()
```

```python
gas_per_gallon_df[['Year2', 'Month']] = gas_per_gallon_df['Label'].str.split(' ', expand=True)
gas_per_gallon_df = gas_per_gallon_df.drop(columns=['Series ID', 'Period', 'Year2', 'Label'])
gas_per_gallon_df.head()
```

```python
gas_per_gal_df = gas_per_gallon_df[["Month", "Year", "Value" ]].copy()
gas_per_gal_df.head()
```

```python
connection_string = f"postgres:{post_pass}@localhost:5432/Project3"
engine = create_engine(f'postgresql://{connection_string}') 
```

```python
gas_per_gal_df.rename(columns={'Month': 'Month', 'Year': 'Year','Value': 'Price'}, inplace=True)
gas_per_gal_df
```

```python
engine.table_names()
```

```python
#add data to sql table
gas_per_gal_df.to_sql(name='gas_price_per_gallon', con=engine, if_exists='append', index=False)
```
