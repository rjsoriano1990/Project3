```python
import os
import pandas as pd
from sqlalchemy import create_engine
from config import post_pass
```

```python
csv_file = os.path.join("Resources/US_gas_prices_region.csv")
```

```python
price_per_gal_reg_df = pd.read_csv(csv_file)
price_per_gal_reg_df.head()
```

```python
# convert "Date" colimn to datetime format

price_per_gal_reg_df['Date'] = pd.to_datetime(price_per_gal_reg_df['Date'])

price_per_gal_reg_df.head()
```

```python
# drop all data prior to 11/23/2011

new_gas_df = price_per_gal_reg_df.loc[(price_per_gal_reg_df['Date'] >= '23-11-2011')]
```

```python
new_gas_df.head()
```

```python
# drop unnessecary columns
# new_gas_df = new_gas_df.drop(new_gas_df.columns[[9]], axis=1)
```

```python
new_gas_df = new_gas_df.rename(columns={'Date': 'date',
    'East Coast': 'east_coast',
    'New England': 'new_england',
    'Central Atlantic': 'central_atlantic',
    'Lower Atlantic': 'lower_atlantic',
    'Midwest Gasoline': 'midewest',
    'Gulf Coast': 'gulf_coast',
    'Rocky Mountain': 'rocky_mountains',
    'West Coast': 'west_coast'
})

new_gas_df.head()
```
